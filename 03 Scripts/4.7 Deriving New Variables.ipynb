```python
# Creating new column for busiest_period_of_day
def busiest_period_of_day(row):

  if row['order_hour_of_day'] in [23, 6, 0, 1, 5, 2, 4, 3]:
    return 'Fewest orders'
  elif row['order_hour_of_day'] in [8, 18, 19, 20, 7, 21, 22]:
    return 'Average orders'
  elif row['order_hour_of_day'] in [10, 11, 14, 15, 13, 12, 16, 9, 17]:
    return 'Most orders'
  else: return 'Not enough data'
```

```python
# Applying the function
ords_prods_merge['busiest_period_of_day'] = ords_prods_merge.apply(busiest_period_of_day, axis = 1)
```

```python
# Checking frequency
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

```python
# Checking order_hour_of_day vs busiest_period_of_day
ords_prods_merge[['order_hour_of_day', 'busiest_period_of_day']].value_counts(dropna=False)
```
